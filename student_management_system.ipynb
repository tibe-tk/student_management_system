```python
def menu():
    print("\n***********STUDENT MANAGEMENT SYSTEM***********")
    print("1. Registration")
    print("2. Search Student")
    print("3. Update Student GPA")
    print("4. Display All Students")
    print("5. Update Student Name")
    print("6. Delete Student Information")
    print("7. Total Number of Students")
    print("8. Total Number of Students based on Gender")
    print("9. Top Scoring Students")
    print("10. Top Scoring Women Students")
    print("11. Search Students based on GPA")
    print("12. Display Frequent Student Name")
    print("13. Display Student per Department")
    print("14. EXIT")

    choice_functions = {
        1: register_student,
        2: search_student,
        3: update_gpa,
        4: show_all,
        5: update_name,
        6: delete_info,
        7: count_students,
        8: count_students_gender,
        9: top_scorer_department,
        10: top_female_scorer_department,
        11: gpa_threshold,
        12: frequent_name,
        13: students_per_department,
        14: exit_program
    }

    try:
        choice = int(input("\nEnter your choice: "))
        selected_function = choice_functions.get(choice)
        if selected_function:
            selected_function()
        else:
            print("Invalid choice!")
            menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def register_student():
    try:
        print("Registration form")
        num_students = int(input("Enter the number of students: "))
        for i in range(1, num_students + 1):
            with open("student_db.txt", 'r') as file:
                st_id = input("Enter student id: ")
                for line in file:
                    stid, name, gpa, dep, gen = line.split(";")
                    while stid == st_id:
                        print("The ID already exists, enter a new ID...")
                        st_id = input("Enter student id: ")
                st_name = input("Enter student name: ")
                while not st_name.isalpha():
                    print("Please enter a valid name...")
                    st_name = input("Enter student name: ")
                st_gpa = input("Enter student GPA: ")
                while not st_gpa.replace('.', '', 1).isdigit() or not 0 <= float(st_gpa) <= 4:
                    print("Please enter a valid GPA...")
                    st_gpa = input("Enter student GPA: ")
                st_dep = input("Enter the department: ")
                while not st_dep.isalpha():
                    print("Please enter a valid department...")
                    st_dep = input("Enter the department: ")
                st_gen = input("Enter student gender m or f: ").lower()
                while st_gen not in ['m', 'f']:
                    print("Please enter 'm' for male or 'f' for female...")
                    st_gen = input("Enter student gender m or f: ").lower()
                with open("student_db.txt", "a") as wfile:
                    wfile.write(f"{st_id};{st_name};{st_gen};{st_dep};{st_gpa}\n")
                    print("Registered successfully. Continue if any...")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def search_student():
    try:
        found = False
        id_search = input("Enter student ID to search: ")
        with open("student_db.txt", 'r+') as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if stid == id_search:
                    print(f"\n ID: {stid}\n Name: {name}\n Gender: {gender}\n Department: {dep}\n GPA: {gpa}")
                    found = True
                    break
        if not found:
            print("Student does not exist")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def update_gpa():
    try:
        st_id = input("Enter student ID: ")
        updated = False
        lines = []
        with open("student_db.txt", 'r') as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if st_id == stid:
                    st_gpa = input("Enter student GPA: ")
                    while not st_gpa.replace('.', '', 1).isdigit() or not 0 <= float(st_gpa) <= 4:
                        print("Please enter a valid GPA...")
                        st_gpa = input("Enter student GPA: ")
                    line = f"{stid};{name};{gender};{dep};{st_gpa}\n"
                    updated = True
                    print(f"Data of student {name} updated successfully!")
                lines.append(line)
        if not updated:
            print(f"Student with ID {st_id} not found.")
        else:
            with open("student_db.txt", 'w') as file:
                file.writelines(lines)
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def show_all():
    try:
        print("Student database:")
        with open("student_db.txt", 'r') as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                print(f"\n ID: {stid} \n Name: {name} \n Gender: {gender} \n Department: {dep} \n GPA: {gpa}")
        menu()
    except Exception as e:
       print("Something went wrong",e)
       menu()
```

```python
def update_name():
    try:
        st_id = input("Enter student ID: ")
        updated = False
        lines = []
        with open("student_db.txt", 'r') as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if st_id == stid:
                    new_name = input("Enter corrected New Name: ")
                    line = f"{stid};{new_name};{gender};{dep};{gpa}"
                    updated = True
                    print(f"Data of student {st_id} updated successfully!")
                lines.append(line)
        if not updated:
            print(f"Student with ID {st_id} not found.")
        else:
            with open("student_db.txt", 'w') as file:
                file.writelines(lines)
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def delete_info():
    try:
        st_id = input("Enter student ID: ")
        deleted = False
        lines = []
        with open("student_db.txt", 'r') as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if st_id != stid:
                    line = f"{stid};{name};{gender};{dep};{gpa}"
                    lines.append(line)
                else:
                    deleted = True
        if not deleted:
            print(f"Student with ID {st_id} not found.")
        else:
            with open("student_db.txt", 'w') as file:
                file.writelines(lines)
            print(f"Student with ID {st_id} is delete.")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def count_students():
    try:
        count=0
        with open("student_db.txt","r") as file:
            for line in file:
                count+=1
        print(f"Total number of students: {count}")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def count_students_gender():
    try:
        with open("student_db.txt","r") as file:
            mcount=0
            fcount=0
            for line in file:
                stid,name,gender,dep,gpa = line.split(";")
                if gender=="m":
                    mcount+=1
                if gender=="f":
                    fcount+=1
            print(f"Total number of male students: {mcount}")
            print(f"Total number of female students: {fcount}")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def top_scorer_department():
    try:
        with open("student_db.txt", "r") as file:
            st_dep = input("Enter the department: ")
            max_gpa = 0
            max_name = ""
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if dep == st_dep and float(gpa) > max_gpa:
                    max_gpa = float(gpa)
                    max_name = name
            print(f"Top scoring student in {st_dep} department: {max_name} with GPA {max_gpa}")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def top_female_scorer_department():
    try:
        with open("student_db.txt", "r") as file:
            st_dep = input("Enter the department: ")
            max_gpa = 0
            max_name = ""
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if dep == st_dep and float(gpa) > max_gpa and gender=="f":
                    max_gpa = float(gpa)
                    max_name = name
            print(f"Top scoring female student in {st_dep} department is: {max_name} with GPA {max_gpa}")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def gpa_threshold():
    try:
        with open("student_db.txt", "r") as file:
            st_gpa = input("Enter the gpa: ")
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if float(gpa) > float(st_gpa):
                  print(f"Top scoring students that greater than {st_gpa} are {name} with gpa {gpa}\n")
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def frequent_name():
    try:
        student_frequency={}
        with open("student_db.txt", "r") as file:
            for line in file:
                stid, name, gender, dep, gpa = line.split(";")
                if name in student_frequency:
                    student_frequency[name] += 1
                else:
                  student_frequency[name] = 1
            for name,count in student_frequency.items():
                print( f"The frequent name is {name} is registered {count} times\n" )
        menu()
    except Exception as e:
       print("Something went wrong")
       menu()
```

```python
def students_per_department():
  try:
    student_frequency={}
    with open("student_db.txt", "r") as file:
        for line in file:
            stid, name, gender, dep, gpa = line.split(";")
            if dep in student_frequency:
              student_frequency[dep] += 1
            else:
               student_frequency[dep] = 1
        for dep,count in student_frequency.items():
          print( f"There is {count} students in {dep} departement\n" )
    menu()
  except Exception as e:
    print("Something went wrong")
    menu()
```
